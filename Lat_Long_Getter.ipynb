### This Notebook is for getting the lats/longs of a bus route

```python
import pandas as pd
import geopandas as gpd


pd.set_option('display.max_columns', None) ### This line makes all the columns display, rather than ellipses shorten
```

```python
### import files 

king_county_bus_stops = pd.read_csv("C:/Users/aaron/Fun Mapping Work/Data/Bus_Stop_Data/Transit_Stops_for_King_County_Metro___transitstop_point.csv")

king_county_bus_lines = gpd.read_file("C:/Users/aaron/Fun Mapping Work/Data/Bus_Stop_Data/Transit_Routes_for_King_County_Metro___transitroute_line")
```

```python
king_county_bus_stops
```

```python
## trim a bit
king_county_bus_stops = king_county_bus_stops[king_county_bus_stops['CURRENT_NEXT_CODE'] == 'CURRENT']
```

```python
king_county_bus_stops
```

```python
stops_gdf = gpd.GeoDataFrame(
    king_county_bus_stops, geometry = gpd.points_from_xy(king_county_bus_stops.X, king_county_bus_stops.Y), crs = "EPSG:2926")
```

```python
route_8_line = king_county_bus_lines[king_county_bus_lines['ROUTE_NUM'] == 8]

king_county_bus_lines.head()
```

```python
### make a spatial join to get all the stops within a close distance of the 8 line.

stops_near_8 = gpd.sjoin(stops_gdf, route_8_line, how = 'inner', predicate='dwithin', distance=50)
```

```python
### I'm done with this BS i am just going to manaully get the stop numbers and get their lat longs/

route_8_stop_numbers = [2672, 2180, 2200, 2244, 2245, 2250, 2255, 2260, 2262, 2264, 29268, 29270, 29277, 29273, 29274, 29275, 
                        29276, 12378, 12380, 43706, 43710, 43712, 43714, 43716, 43720, 27335, 12630, 12010, 43724,
                        43728, 12479, 12487, 12484, 12486, 8402] # length 29
```

```python
route_8_eastbound = stops_gdf[stops_gdf['STOP_ID'].isin(route_8_stop_numbers)]
```

```python
route_8_eastbound.explore()
```

```python
### reprojecting to 4326 to get the lats/longs

route_8_projected = route_8_eastbound.to_crs(4326)
```

```python
point_list = route_8_projected.geometry
```
